## What is the Output?

```python
s = 'Final Exam'
for c in s:
    if c == ' ':
        print('?'*5, end='')  ## end = empty, next output willl not use newlines
    elif c.isalpha():
        print(str(c).upper(), end='')
    else:
        print(c, end='', )

```

```python
s = 'Final Exam'
for c in s:
    if c == ' ':
        continue ## skip the space 
    elif c.isalpha():
        print(str(c).lower(), end='')
   
```

```python
s = 'Final-Exam'
for i,c in enumerate(s):
    if c == '-':
        break              ## exit loop 
    elif c.isalpha() & i %2 == 0:  ## even numb -> upper case
        print(str(c).upper(), end='')
    elif c.isalpha() & i %2 == 1 :  ## odd numb -> lower case
        print(str(c).lower(), end='')
        
```

```python
sql_stmt = 'SELECT genre,AVG(duration) FROM MOVIES GROUP BY genre'
rows = cursor.execute(sql_stmt)
pd.DataFrame(rows, columns=[x[0] for x in cursor.description])
```
