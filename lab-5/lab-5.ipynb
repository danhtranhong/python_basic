Define a function with a name <strong>get_properties()</strong> that takes a <strong>NumPy</strong> array as parameters and returns a <strong>dictionary</strong> that includes the following array properties as key-value pairs: (<strong>8 Marks</strong>)
<ul>
    <li> <strong>shape</strong>
    <li> <strong>dimension</strong>
    <li> <strong>size</strong>
   <li> element <strong>type</strong>
</ul>

```python
# your code goes here 
def get_properties(item):
    #print(f"get_shape = {item}.shape")
    get_shape = item.shape
    get_dimension = item.ndim
    get_size = item.size
    get_type = item.dtype
    image_dictionary = { 'Shape': get_shape, 'Dimension':get_dimension, 'Size': get_size, 'Type': get_type}
    print(type(image_dictionary))
    return image_dictionary

print(get_properties(grayscale_np_array))
print(get_properties(img_np_array))
```

Using the function <strong>get_properties()</strong>, print the properties of both <strong>image NumPy arrays</strong> in the following format: (<strong>8 Marks</strong>)
<pre>
Color Image NumPy Array Properties
----------------------------------
{('Shape', (500, 421, 3))}
{('Dimension', 3)}
{('Size', 631500)}
{('Type', dtype('uint8'))}
Grayscale Image NumPy Array Properties
--------------------------------------
{('Shape', (500, 421))}
{('Dimension', 2)}
{('Size', 210500)}
{('Type', dtype('uint8'))}
</pre>

```python
# your code goes here 
def get_properties(item,colour='yes'):
    grayscale_title = "Grayscale Image NumPy Array Properties"
    colour_title = "Color Image NumPy Array Properties"
    if colour == "yes":
        ruler = '-'*(len(colour_title))
        print(f"{colour_title}\n{ruler}")
        colour_properties = {
        'Shape': item.shape,
        'Dimension': item.ndim,
        'Size': item.size,
        'Type': item.dtype
        }
        for key, value in colour_properties.items():
           print(f"{{('{key}',{repr(value)})}}")
        
    else:
        ruler = '-'*(len(grayscale_title))
        print(f"{grayscale_title}\n{ruler}")
        gray_properties = {
        'Shape': item.shape,
        'Dimension': item.ndim,
        'Size': item.size,
        'Type': item.dtype
        }
        for key, value in gray_properties.items():
           print(f"{{('{key}',{repr(value)})}}")

get_properties(img_np_array)
get_properties(grayscale_np_array,colour='no')
```
