```python
# you code goes here
plt.figure(figsize=(10, 10))
plt.scatter(drinks_2.index, drinks_2['total_litres_of_pure_alcohol'], marker='H',c='purple')
plt.style.use("fast")
# Add labels and title
plt.xlabel('Country index')
plt.ylabel('Total Litres of Pure Alcohol')
plt.title('Scatter Plot of Total Litres of Pure Alcohol')

plt.show()

plt.savefig(r'.\results\scatter.png')

```

<hr>
Extract the top 10 countries with the highest alcohol consumption.<strong>(5 Marks)</strong>
<hr>

```python
# you code goes here
drinks_2_filter =  drinks_2.sort_values('total_litres_of_pure_alcohol',ascending=False).head(10)
drinks_2_filter
```

<hr>
Produce the following figure using the <strong>fast</strong> style and save with a name of <strong>line.png</strong>. All figures must be saved in the <strong>results</strong> directory. <strong>(5 Marks)</strong>
<figure>
  <img src="figures/line.png" alt="Line Plot" style="width:100%">
  <figcaption>Fig.2 - Line Plot.</figcaption>
</figure>
<hr>

```python
# you code goes here
y = drinks_2_filter['total_litres_of_pure_alcohol']
x = drinks_2_filter['country']
plt.figure(figsize=(10, 10))
plt.grid()
plt.title('top 10 countries with highest alcohol consumption')
plt.xlabel('Countries')
plt.ylabel('Pure Alcohol litter')

plt.style.use('fast')
plt.xticks(rotation=45)
plt.plot(x,y,color='c',linestyle='--',marker='o')

plt.savefig(r'.\results\line.png')
```

```python
# you code goes here
y = drinks_2_filter['total_litres_of_pure_alcohol']
x = drinks_2_filter['country']
plt.figure(figsize=(12,8))
plt.grid()
plt.title('top 10 countries with highest alcohol consumption')
plt.xlabel('Countries')
plt.ylabel('Pure Alcohol litter')

plt.style.use('classic')
plt.xticks(rotation=45)
plt.bar(x,y,color='#B4EEB4')

plt.savefig(r'.\results\line.png')
plt.show()
```

Produce the following figure using the <strong>Solarize_Light2</strong> style and save with a name of <strong>boxplot.png</strong>. All figures must be saved in the <strong>results</strong> directory. <strong>(5 Marks)</strong>

```python
# you code goes here

#x = drinks_2_filter['country']
y = drinks['wine_servings']
plt.figure(figsize=(15,10))
plt.style.use('Solarize_Light2')

plt.boxplot(y,boxprops=dict(linewidth=2.0, color='blue'),whiskerprops=dict(linestyle='--',linewidth=2.0, color='blue'),medianprops=dict(linewidth=2.0,color='red'))

plt.savefig(r'.\results\boxplot.png')

#drinks_2.boxplot(boxprops= dict(linewidth=1.0, color='blue'), whiskerprops=dict(linestyle='--',linewidth=1.0, color='blue'))
plt.show()

```
