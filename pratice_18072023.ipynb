```python
import matplotlib as plt
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline

x = np.linspace(start=-10, stop=10,num=11)
y= x**3
display(x,y)
plt.plot(x,y)
```

```python
plt.title(r'$y=x^3$')
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(x,y,color='b',linestyle='--',marker='o')
```
